```python
import cv2
import tkinter as tk
from tkinter import *
from tkinter import filedialog
from tkinter.filedialog import askopenfile
from tkinter.filedialog import asksaveasfile
import os

tkWindow = Tk()  
tkWindow.geometry('600x450')  
tkWindow.title('Image compression')
tkWindow['bg'] = '#FFB6C1'
my_font1=('times', 18, 'bold')

HEIGHT = 400
WIDTH = 500

Label(tkWindow, text='Choose a compression type:', pady=40, bg='#FFB6C1', font=my_font1).pack()


def showLossless(): 
    
    Window = tk.Toplevel()
    canvas = tk.Canvas(Window, height=HEIGHT, width=WIDTH)
    canvas['bg'] = '#FFB6C1'
    canvas.pack()
    Label(canvas, text='Upload a .BMP image!', pady=20, bg='#FFB6C1', font=my_font1).pack()
    Label(canvas, text='Then save the image with desired extension.', pady=20, bg='#FFB6C1', font=my_font1).pack()

    def upload_file():
        filename = filedialog.askopenfilename()
        img = cv2.imread(filename) 
        cv2.imshow('BMP initial image', img)
        
        files = [('Png Files', '*.png*'), ('Tiff Document', '*.tiff')]
        file = asksaveasfile(filetypes = files, defaultextension = files)
        
        cv2.imwrite(file.name, img, [int(cv2.IMWRITE_PNG_COMPRESSION), 9])
        img = cv2.imread(file.name)
        cv2.imshow('Converted image with smaller size', img)
        Label(canvas, text='You can see the lossless compression effects:', pady=20, bg='#FFB6C1').pack()
        Label(canvas, text="Size of initial file is {} bytes.".format(os.stat(filename).st_size), pady=10, bg='#FFB6C1', fg="red").pack()
        Label(canvas, text="Size of compressed file is {} bytes.".format(os.stat(file.name).st_size), pady=10, bg='#FFB6C1', fg="red").pack()

    Button(canvas, text="Upload", padx=10, pady=5, command=upload_file).pack()


def showLossy():  
    
    Window = tk.Toplevel()
    canvas = tk.Canvas(Window, height=HEIGHT, width=WIDTH)
    canvas['bg'] = '#FFB6C1'
    canvas.pack()
    
    Label(canvas, text='Upload a .BMP image!', pady=20, bg='#FFB6C1', font=my_font1).pack()

    
    def upload_file():
        filename = filedialog.askopenfilename()
        img = cv2.imread(filename)
        cv2.imshow('BMP initial image', img)
        
        cv2.imwrite('jpg_100.jpeg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 100])
        img_jpg100 = cv2.imread('jpg_100.jpeg')
        cv2.imshow('JPEG higher quality', img_jpg100)
        cv2.imwrite('jpg_0.jpeg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 0])
        img_jpg0 = cv2.imread('jpg_0.jpeg')
        cv2.imshow('JPEG lower quality', img_jpg0)
    
    Button(canvas, text="Upload", padx=10, pady=5, command=upload_file).pack()

    
    

button1 = Button(tkWindow, text = 'Lossy', command = showLossy)  
button1.pack()  

button2 = Button(tkWindow, text = 'Lossless', command = showLossless)  
button2.pack()  

    
tkWindow.mainloop()
```
